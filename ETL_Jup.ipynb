```python
#!pip install psycopg2
#Import libraries

import pandas as pd
from sqlalchemy import create_engine,Column, Integer, String
from sqlalchemy.ext.declarative import declarative_base
import psycopg2
Base = declarative_base()
```

```python

class Location(Base):
    __tablename__ = 'denver_demo_trees'
    Neighborhood_id = Column(Integer, primary_key=True)
    Neighborhood=Column(String)
    Population=Column(Integer) 
    Hispanic=Column(Integer)
    White=Column(Integer)
    Black=Column(Integer)
    Native_AM=Column(Integer)
    Asian=Column(Integer)
    Hawaiian_Pacific=Column(Integer)
    Other=Column(Integer)
    Tree_Species=Column(String)
    Tree_Diameter=Column(String)
    


   
# Base.metadata.drop_all(engine)
```

```python
#read demographics csv data file

demo_file = "demographics.csv"
demo_df = pd.read_csv(demo_file)
demo_df
```

```python
#demographics clean up


#print list of columns to find count
demo_df.columns

#drop columns 
demo_df.drop(demo_df.iloc[:, 10:127], inplace = True, axis = 1) 

#confirm all columns have the same number of items
demo_df.count()

#rename columns
demographics = demo_df.rename(columns={'NBHD_ID':'Neighborhood_id','NBRHD_NAME':'Neighborhood',
                                      'POPULATION_2010':'Population',
                                        'HISPANIC_2010':'Hispanic',
                                      'WHITE_2010':'White',
                                      'BLACK_2010':'Black',
                                      'NATIVEAM_2010':'Native_AM',
                                        'ASIAN_2010': 'Asian',
                                       'HAWPACIS_2010':'Hawaiian_Pacific',
                                       'OTHER_2010':'Other'
                                      })
#duplicate/similar neighborhood names?

demographics.columns
demographics['Neighborhood'].unique()
```

```python
#read trees csv data file

tree_file = "tree_inventory.csv"
tree_df = pd.read_csv(tree_file)
tree_df

tree_df.drop_duplicates('SITE_ID')
```

```python
#print list of tree columns

tree_df.columns

# #drop columns
tree_inv = tree_df[['SPECIES_BOTANIC', 'DIAMETER','NEIGHBOR']]
tree_inv

#rename columns
tree_inventory = tree_inv.rename(columns={'SPECIES_BOTANIC':'Tree_Species',
                                          'DIAMETER':'Tree_Diameter','NEIGHBOR':'Neighborhood'})
tree_inventory

#check for blanks
tree_inventory.count()



#duplicate/similar neighborhood names?

tree_inventory.columns
tree_inventory['Neighborhood'].unique()



```

```python
#merge

merge_df = demographics.merge(tree_inventory, on='Neighborhood')
merge_df

#change index
merge_df.set_index('Neighborhood_id', inplace=True)
merge_df


merge_df
```

```python
merge_df.to_sql(name='merge_df', con=engine, if_exists='append', index=False)
```
